## Matriz de similitud

```python
data
```

```python
inicial = np.array([0,1,4])
fin = np.arange(36, data.shape[1])
dat_a = np.append(inicial, fin)
```

```python
data_a = data.iloc[:, dat_a]
data_a
```

```python
data_sim = data_a.iloc[:,3:]
data_sim
```

```python
import scipy.spatial.distance as sc
```

```python
# Medida de similitud para datos binarios Jaccard
sc.pdist(data_sim, 'jaccard')
```

```python
def pdistance(df, metric):
    dist = sc.pdist(df, metric)
    #Matriz de distancias de similitud
    DIST = pd.DataFrame(sc.squareform(dist))
    
    return DIST
```

```python
# Matriz de similidtud Jaccard
pdistance(data_sim, 'jaccard')
```

```python
# Euclideana
eu = sc.pdist(data_sim, 'euclidean')
eu = sc.squareform(eu)
eu
```

```python
# Matriz de similidtud euclidean
pdistance(data_sim, 'euclidean')
```

```python
# Canberra
can = sc.pdist(data_sim, 'canberra')
can
```

```python
# Matriz de similidtud Canberra
pdistance(data_sim, 'canberra')
```

```python
z = (data_sim - data_sim.mean()) / data_sim.std()
z
```
